```python
# --- 1. Importar bibliotecas ---
# pip install ...

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python

# --- 2. Carregar os dados CSV ---
users = pd.read_csv("archive/users.csv")
campaigns = pd.read_csv("archive/campaigns.csv")
ads = pd.read_csv("archive/ads.csv")
ad_events = pd.read_csv("archive/ad_events.csv")
```

```python
# --- 3. Primeira visÃ£o dos dados ---
print("ðŸ“Š Tamanho das tabelas:")
print("Users:", users.shape)
print("Campaigns:", campaigns.shape)
print("Ads:", ads.shape)
print("Ad Events:", ad_events.shape)
```

```python
# Visualizar primeiras linhas
display(users.head())
display(campaigns.head())
display(ads.head())
display(ad_events.head())
```

```python
# --- 4. Estrutura e valores ausentes ---
# aparentemente nÃ£o tem dados nulos

for name, df in zip(["Users", "Campaigns", "Ads", "Ad Events"], [users, campaigns, ads, ad_events]):
    print(f"\n--- {name} ---")
    print(df.info())
    print("Valores nulos:\n", df.isna().sum())
```
