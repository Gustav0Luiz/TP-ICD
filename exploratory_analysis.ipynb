## Python Exploratory Analysis

```python
# --- 1. Importar bibliotecas ---
# pip install ...

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# --- 2. Carregar os dados CSV ---

pokemon = pd.read_csv("archive/pokemon.csv")
```

```python
# --- 3. Primeira visÃ£o dos dados ---
print("ðŸ“Š Tamanho das tabelas:")
print("Table (lines,cols):", pokemon.shape)
```

```python
# Visualizar primeiras linhas
display(pokemon.head())
```

```python
# --- 4. DistribuiÃ§Ã£o por geraÃ§Ã£o ---
plt.figure(figsize=(5,3))
sns.countplot(data=pokemon, x="generation")
plt.title("DistribuiÃ§Ã£o de PokÃ©mons por GeraÃ§Ã£o")
plt.show()
```

```python
# --- 5. DistribuiÃ§Ã£o de tipos primÃ¡rios ---
plt.figure(figsize=(5,3))
sns.countplot(data=pokemon, x="type1", order=pokemon["type1"].value_counts().index,)
plt.title("DistribuiÃ§Ã£o dos Tipos PrimÃ¡rios de PokÃ©mon")
plt.xticks(rotation=45)
plt.show()
```

```python
# --- 6. MÃ©dia de ataque por tipo ---

# Calcular a mÃ©dia de ataque por tipo primÃ¡rio para todas as geraÃ§Ãµes
attack_mean_by_type = pokemon.groupby("type1")["attack"].mean().reset_index()

# Ordenar do maior para o menor
attack_mean_by_type = attack_mean_by_type.sort_values(by="attack", ascending=False)

# Plotar
plt.figure(figsize=(5, 3)) # Aumentei o tamanho do grÃ¡fico para acomodar mais tipos
sns.barplot(data=attack_mean_by_type, x="type1", y="attack")
plt.title("MÃ©dia de Ataque por Tipo (Todas as GeraÃ§Ãµes)")
plt.ylabel("MÃ©dia de Ataque")
plt.xlabel("Tipo PrimÃ¡rio")
plt.xticks(rotation=45)
plt.show()
```
